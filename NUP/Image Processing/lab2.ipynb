## STEP 3

```python
histogram, bins = np.histogram(gray_image.flatten(), 256, [0, 256])
pdf = histogram.cumsum()
cdf = pdf * histogram.max() / pdf.max()
```

```python
plt.plot(cdf, color="black")
plt.title("Original Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Cumulative Frequency")
```

## STEP 4

```python
equalized_image = cv2.equalizeHist(gray_image)
equalized_histogram, _ = np.histogram(equalized_image.flatten(), 256, [0, 256])
equalized_pdf = equalized_histogram.cumsum()
equalized_cdf = equalized_pdf * equalized_histogram.max() / equalized_pdf.max()
```

```python
plt.imshow(equalized_image, cmap="gray")
plt.title("Equalized Image")
plt.axis("off")
```

```python
plt.figure(figsize=(6, 4))
plt.plot(equalized_cdf, color="black")
plt.title("Equalized Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Cumulative Frequency")
```

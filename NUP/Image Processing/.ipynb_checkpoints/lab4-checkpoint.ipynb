## Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

## STEP 1

```python
image = cv2.imread("lena.jpg", 0)
```

## STEP 2

```python
mean = 0
std_dev = 0.1
gaussian_noise = np.random.normal(mean, std_dev, image.shape)
noisy_image = np.clip(image + gaussian_noise, 0, 255).astype(np.uint8)
```

## STEP 3

```python
mean_3x3 = ndimage.convolve(image, np.ones((3, 3)) / 9)
mean_7x7 = ndimage.convolve(image, np.ones((7, 7)) / 49)
```

## STEP 4

```python
salt_pepper_ratio = 0.05
salt_pepper_noise = np.random.choice([0, 255], size=image.shape, p=[1 - salt_pepper_ratio, salt_pepper_ratio])
noisy_image_sp = np.where(salt_pepper_noise == 255, 255, np.where(salt_pepper_noise == 0, 0, image))
```

## STEP 5

```python
median_3x3 = ndimage.median_filter(noisy_image_sp, size=3)
median_7x7 = ndimage.median_filter(noisy_image_sp, size=7)
```

## STEP 6

```python
plt.figure(figsize=(10, 6))

plt.subplot(221)
plt.imshow(mean_3x3, cmap="gray")
plt.title("Mean (3x3)")
plt.axis("off")

plt.subplot(222)
plt.imshow(mean_7x7, cmap="gray")
plt.title("Mean (7x7)")
plt.axis("off")

plt.subplot(223)
plt.imshow(median_3x3, cmap="gray")
plt.title("Median (3x3)")
plt.axis("off")

plt.subplot(224)
plt.imshow(median_7x7, cmap="gray")
plt.title("Median (7x7)")
plt.axis("off")

plt.tight_layout()
plt.show()
```

## STEP 7

```python
f = np.fft.fft2(image)
magnitude = np.abs(f)
phase = np.angle(f)
```

## STEP 8

```python
mask = np.zeros_like(f)
rows, cols = image.shape
crow, ccol = rows // 2, cols // 2
mask[crow-2:crow+3, ccol-2:ccol+3] = 1
filtered_f = f * mask
```

## STEP 9

```python
unsharp_mask = np.abs(np.fft.ifft2(filtered_f)) - image
```

## STEP 10

```python
sharpened_image = np.clip(image + unsharp_mask, 0, 255).astype(np.uint8)
```

## STEP 11

```python
plt.figure(figsize=(10, 6))

plt.subplot(211)
plt.imshow(np.log(1 + magnitude), cmap="gray")
plt.title("Magnitude Spectrum")
plt.axis("off")

plt.subplot(212)
plt.imshow(np.log(1 + np.abs(filtered_f)), cmap="gray")
plt.title("Filtered Spectrum")
plt.axis("off")

plt.tight_layout()
plt.show()
```
